```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=pd.DataFrame({"Month":["jan","feb","march","april"],
                     "Week1":[7000,9000,10500,11750],
                     "Week2":[8000,9200,12000,15000],
                     "Week3":[10000,12000,11000,1750],
                     "Week4":[1000,2000,3000,1500]})
print(data)
```

```python
df=pd.melt(data,id_vars="Month",var_name="Week",value_name="Expense")
print(df)
```

```python
sns.barplot(x="Month",y="Expense",data=df,hue="Week",palette="light:teal",width=2,edgecolor="yellow",linewidth=2,orient="v")
plt.title("Monthly Expense")
```

```python
Monthly_expense=df.groupby("Month")["Expense"].sum()
print(Monthly_expense)
```

```python
plt.pie(Monthly_expense,labels=Monthly_expense.index,
        autopct='%1.0f%%',
        wedgeprops={'edgecolor':'black','linestyle':'-','linewidth':2},
        colors=['yellow','green','red','blue'])
plt.show()
```
